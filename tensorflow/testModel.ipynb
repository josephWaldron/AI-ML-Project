```python
#load model
from tensorflow.keras.models import load_model
model = load_model("../models/tf_inception3.h5")

#load a random image
import numpy as np
from tensorflow.keras.preprocessing import image
import random
import os
import matplotlib.pyplot as plt

#test dir
test_dir = "../data/CUB_200_2011/images/"
classes = os.listdir(test_dir)
# random class
class_name = random.choice(classes)
# random image
img_name = random.choice(os.listdir(test_dir + class_name))
# load image
img = image.load_img(test_dir + class_name + "/" + img_name, target_size=(224, 224))
# convert image to array
img_array = image.img_to_array(img)
# expand dimension of image
img_expand = np.expand_dims(img_array, axis=0)
# predict image
pred = model.predict(img_expand)
print(pred)
#get the top 3 classes
top3 = np.argsort(pred[0])[:-4:-1]
for i in range(3):
    print("{}".format(classes[top3[i]])+" ({:.3})".format(pred[0][top3[i]]))
pred_class = classes[np.argmax(pred)]
# plot image
plt.imshow(img)
plt.title("Class: " + class_name + " | " + "Prediction: " + pred_class)
plt.show()
```

```python
#load model
from tensorflow.keras.models import load_model
model_inception2 = load_model("../models/tf_inception2.h5")

#load a random image
import numpy as np
from tensorflow.keras.preprocessing import image
import random
import os
import matplotlib.pyplot as plt

#test dir
test_dir = "../data/test/"
classes = os.listdir(test_dir)

n = 10  # number of images to classify

for i in range(n):
    # random class
    class_name = random.choice(classes)
    # random image
    img_name = random.choice(os.listdir(test_dir + class_name))
    # load image
    img = image.load_img(test_dir + class_name + "/" + img_name, target_size=(224, 224))
    # convert image to array
    img_array = image.img_to_array(img)
    # expand dimension of image
    img_expand = np.expand_dims(img_array, axis=0)
    # predict image
    pred = model_inception2.predict(img_expand)
    # get class name
    pred_class = classes[np.argmax(pred)]
    # plot image
    plt.imshow(img)
    plt.title("Class: " + class_name + " | " + "Prediction: " + pred_class)
    plt.show()
```
